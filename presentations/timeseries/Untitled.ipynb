```python
def collatz(n: int) -> int:
    if n == 1:
        return 1
    elif n % 2 == 0:
        return 1 + collatz((n) // 2)
    else:  # n % 2 == 1:
        return 1 + collatz(3 * (n) + 1 )
```

```python
collatz(7)
```

```python
def collatz_sequence(n):
    num = n
    while True:
        yield num
        if num == 1:
            break
        if num % 2 == 0:
            num = num // 2
        else:
            num = 3*num + 1
```

```python
import seaborn as sns

Y = list(collatz_sequence(257))

X = range(len(Y))

sns.lineplot(x=X,y=Y)
```

```python
def collatz(n: int) -> int:
    if n == 1:
        return 1
    elif n % 2 == 0:
        return 1 + collatz((n) // 2)
    else:  # n % 2 == 1:
        return 1 + collatz(3 * (n) + 1 )
```

```python
collatz(7)
```

```python
def collatz_sequence(n):
    num = n
    while True:
        yield num
        if num == 1:
            break
        if num % 2 == 0:
            num = num // 2
        else:
            num = 3*num + 1
```

```python
import seaborn as sns

Y = list(collatz_sequence(871))

X = range(len(Y))

sns.lineplot(x=X,y=Y)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import darts
from darts.utils.statistics import plot_acf, plot_pacf, check_seasonality
```

```python
ds_coll = pd.Series(Y)
ts_coll = darts.TimeSeries.from_series(ds_coll)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(3,1, figsize=(16, 8))
fig.tight_layout()
ts_coll.plot(ax=ax1)
ax1.set_title("")
ax2.set_xlabel("lag")
plot_acf(ts_coll, axis=ax2)
plot_pacf(ts_coll, axis=ax3)
```

```python
ts_diff = ts_coll.diff()

fig, (ax1, ax2, ax3) = plt.subplots(3,1, figsize=(16, 8))
fig.tight_layout()
ts_diff.plot(ax=ax1)
ax1.set_title("")
ax2.set_xlabel("lag")
plot_acf(ts_diff, axis=ax2)
plot_pacf(ts_coll, axis=ax3)
```
